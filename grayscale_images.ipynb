```python
import cv2
from matplotlib import pyplot as plt
```

```python
def imshow(title="", image=None, size=10):
  w, h= image.shape[1], image.shape[0]
  aspect_ration=w/h
  plt.figure(figsize=(size*aspect_ration, size))
  plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
  plt.title(title)
  plt.show()
```

```python
image = cv2.imread("/content/images/IMG_8295.JPG")
imshow("Image", image)
```
